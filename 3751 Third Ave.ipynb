```python
newSqft           = 7200
oldBuildCost      = 35+50+100
deck              = 100
concretePedaStool = 100
Elevator          = 100
carryCostRate     = 4
purchasePrice     = 1850000
fig = go.Figure()   
df = pd.DataFrame(columns=["Valuation","BuildCost","TotalCost","Profit","ExpRate","rentRoll","Cap","CostPerSqft"])
idx = 0
for expRate in [33]:#range(30,35+1,5):
    for rent in [1700]:#range(1500,1800+1,100):
        for capPF in np.arange(3,5,.5):        
            for costPerSqft in np.arange(200,300+1,25):
                buildCost = newSqft * costPerSqft +oldBuildCost +deck+concretePedaStool+Elevator
                rentRoll = 18 * rent
                valuation = rentRoll * (100-expRate)/100. * 12 / (capPF/100.)
                totalCost = purchasePrice+buildCost + (purchasePrice+buildCost)*carryCostRate/100.
                #print("${Valuation:,}, ${Cost:,}, ${Profit:,}".format(
                #    Valuation=int(valuation), Cost=int(totalCost),Profit=int(valuation-totalCost)))
                df.loc[idx] = [int(valuation),int(buildCost),int(totalCost),int(valuation-totalCost),expRate,rentRoll,capPF,costPerSqft]
                idx+=1
df    
```
