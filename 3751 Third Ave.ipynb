```python
from IPython.display import display, HTML
import plotly
from plotly.graph_objs import graph_objs as go
from plotly.offline import iplot
from ipywidgets import *
```

```python
# new building
newSqft           = 7200
#35k 3bd to 2X1bd1ba, 50k exterior and ~20 for each unit
oldBuildCost      = 35000+50000+100000 
# deck over existing build used by all tenants, over 1500
deck              = 100000
# concrete pedastool for first floor, to allow for future retail
concretePedaStool = 100000
# elevator
Elevator          = 100000
# assuming purchase/construction cost and assuming all tenants vacant
carryCostRate     = 4
purchasePrice     = 1850000
fig = go.Figure()   
df = pd.DataFrame(columns=["Valuation","BuildCost","TotalCost","Profit","ExpRate","rentRoll","Cap","CostPerSqft"])
idx = 0
for expRate in range(30,35+1,1):
    for rent in range(1500,1800+1,100):
        for capPF in np.arange(3,5+.1,.25):        
            for costPerSqft in np.arange(200,300+1,10):
                buildCost = newSqft * costPerSqft +oldBuildCost +deck+concretePedaStool+Elevator
                rentRoll = 18 * rent
                valuation = rentRoll * (100-expRate)/100. * 12 / (capPF/100.)
                totalCost = purchasePrice+buildCost + (purchasePrice+buildCost)*carryCostRate/100.
                #print("${Valuation:,}, ${Cost:,}, ${Profit:,}".format(
                #    Valuation=int(valuation), Cost=int(totalCost),Profit=int(valuation-totalCost)))
                df.loc[idx] = [int(valuation),int(buildCost),int(totalCost),int(valuation-totalCost),expRate,rentRoll,capPF,costPerSqft]
                idx+=1
```

```python
import plotly.express as px

#fig = go.Figure()
#fig.add_trace(go.Scatter(y=df["Valuation"],line = dict(color='green', width=1, dash='dash')))
#fig = px.scatter_3d(df,x="CostPerSqft",y="Cap",z="Valuation")
fig = px.parallel_coordinates(df, color="Profit",
                              dimensions=['CostPerSqft',
                                          'Cap', 
                                          'ExpRate',
                                          'rentRoll',
                                          "Profit"
                                          ])
fig.show()
print("%d" % min(df["Profit"]))
```
